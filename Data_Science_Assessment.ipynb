Imputation

```python
dtype_obj = ["Age", "HoursPerWeek", "TotalHours"]

# Replacing '?' as null value
df[dtype_obj] = df[dtype_obj].apply(lambda x: x.replace("?", np.nan))

# Converting the dtype of these features to numeric
for col in dtype_obj:
    df[col] = pd.to_numeric(df[col])
```

```python
plt.figure(figsize=(15, 10))
sns.boxenplot(data=df, x="LeagueIndex", y="Age", k_depth="trustworthy")
```

```python
#Imputing missing values with median of Age for rank 7
#This is done so because age range decreases with higher league index
med_val = df[df["LeagueIndex"] == 7]["Age"].median()
df["Age"] = df["Age"].fillna(med_val)
```

```python
plt.figure(figsize=(15, 10))
sns.boxenplot(data=df, x="LeagueIndex", y="HoursPerWeek", k_depth="trustworthy")
```

```python
df[df["HoursPerWeek"].isnull()]["LeagueIndex"].unique()
```

```python
median_league_5 = df[df["LeagueIndex"] == 5]["HoursPerWeek"].median()
df.loc[
    (df["HoursPerWeek"].isnull()) & (df["LeagueIndex"] == 5), "HoursPerWeek"
] = median_league_5

# Now imputing the missing value for player with missing hoursperweek for leagueindex = 8
median_league_8 = df[df["LeagueIndex"] == 7]["HoursPerWeek"].median()
df.loc[
    (df["HoursPerWeek"].isnull()) & (df["LeagueIndex"] == 8), "HoursPerWeek"
] = median_league_8
```

```python
sns.boxenplot(data=df, x="LeagueIndex", y="TotalHours", k_depth="trustworthy")
```

```python
df[df["TotalHours"].isnull()]["LeagueIndex"].unique()
```

```python
median_tot_hours = df[df["LeagueIndex"].between(5, 7)]["TotalHours"].median()
df.loc[df["TotalHours"].isnull(), "TotalHours"] = median_tot_hours
print("Total missing values now are {}".format(df.isnull().sum().sum()))
```

### Feature selection and engineering

Looking at multicollinearity and variance inflation

```python
plt.figure(figsize=(20, 15))
corr = df.corr()
sns.heatmap(corr, annot=True)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#To find Correlation of features we find Variance inflation factor of each feature
df1 = df.drop(["LeagueIndex"], axis=1)
vif = pd.DataFrame()
vif["feature"] = df1.columns

# calculating VIF for each feature
vif["VIF"] = [
    variance_inflation_factor(df1.values, i)
    for i in range(len(df1.columns))
]  
```

```python
vif
```

1. Usually VIF >10 suggests high 
multicolinearity between features .  
2. Taking into account VIF and correlation with LeagueIndex we can drop some features.
3. We select the features with very high VIF and check their correlation with LeagueIndex. If there are other features which are correlated with the selected ones and have higher correlation score than the selected ones then the selected ones are dropped from the dataset. 
4. Based on the analysis we remove, ActioninPAC, UniqueUnitsMade, SelectHotkeys, APM, NumberOfPACs, TotalMapExplored, GapBetweenPACs

```python
col_to_drop = [
    "ActionsInPAC",
    "UniqueUnitsMade",
    "SelectByHotkeys",
    "APM",
    "NumberOfPACs",
    "TotalMapExplored",
    "GapBetweenPACs",
]
df.drop(col_to_drop, axis=1, inplace=True)
```

```python
df.columns
```

```python
#To find Correlation of features we find Variance inflation factor of each feature
df1 = df.drop(["LeagueIndex"], axis=1)
vif = pd.DataFrame()
vif["feature"] = df1.columns

vif["VIF"] = [
    variance_inflation_factor(df1.values, i)
    for i in range(len(df1.columns))
]  
```

```python
plt.figure(figsize=(20, 15))
corr = df.corr()
sns.heatmap(corr, annot=True)
```

```python
vif
```

1. We further drop **ComplexUnitsMade** and **HoursPerWeek**. 
2. ComplexUnitsMade is correlated with ComplexAbilitiesUsed and has higher VIF than ComplexAbilitiesUsed.
3. HoursPerWeek has lesser correlation with LeagueIndex than TotalHoursPlayed

```python
df.drop(['ComplexUnitsMade'],axis=1,inplace=True)
```

```python
df.drop(['HoursPerWeek'],axis = 1, inplace=True)
```

```python
Y = df['LeagueIndex']
df_copy = df
df_copy.drop(['LeagueIndex'],axis=1,inplace=True)
X = df_copy
```

Splitting into Train and Test

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.20, random_state=42,stratify = Y)
```

Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler().fit(X_train)
X_train_stand = scale.transform(X_train)
X_test_stand = scale.transform(X_test)
```

```python
rf = RandomForestRegressor(random_state=0)

rf.fit(X_train_stand,y_train)
```

```python
features = X.columns
```

```python
features
```

```python
f_i = list(zip(features,rf.feature_importances_))
f_i.sort(key = lambda x : x[1])
plt.barh([x[0] for x in f_i],[x[1] for x in f_i])

plt.show()
```

```python
#Dropping the Age column as well since its ranked low in the feature importance 
#and also is weakly correlated with LeagueIndex as seen in the correlation diagram
df_copy = df
df_copy.drop(['Age'],axis=1,inplace=True)
X = df_copy
```

```python
f_i = list(zip(features,rf.feature_importances_))
f_i.sort(key = lambda x : x[1])
plt.barh([x[0] for x in f_i],[x[1] for x in f_i])

plt.show()
```

```python
rfe = RFECV(rf,cv=5,scoring="neg_mean_squared_error")

rfe.fit(X_train_stand,y_train)
selected_features = np.array(features)[rfe.get_support()]
```

### Modeling

Using the RandomForestClassifier to model the data

```python
from sklearn.ensemble import RandomForestClassifier
```

X_train_stand and y_train is calculated after dropping the age column

```python
classifier_rf = RandomForestClassifier(random_state=42, n_jobs=-1, max_depth=5,
                                       n_estimators=100, oob_score=True)
classifier_rf.fit(X_train_stand, y_train)
classifier_rf.oob_score_
```

Using GridSearchCV to tune the parameters

```python
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
params = {
    'max_depth': [2,3,5,10,20],
    'min_samples_leaf': [5,10,20,50,100,200],
    'n_estimators': [10,25,30,50,100,200]
}
from sklearn.model_selection import GridSearchCV
# Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf,
                           param_grid=params,
                           cv = 4,
                           n_jobs=-1, verbose=1, scoring="accuracy")
```

```python
grid_search.fit(X_train, y_train)
```

```python
rf_best = grid_search.best_estimator_
rf_best
```

Train accuracy

```python
grid_search.best_score_
```

Test accuracy

```python
grid_search.score(X_test_stand, y_test)
```
